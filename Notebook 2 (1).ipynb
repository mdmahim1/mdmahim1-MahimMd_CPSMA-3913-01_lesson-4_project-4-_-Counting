(a) How many outcomes were possible in your simulation if it is considered a sequence in that particular order?

The permutation is a method of arranging results where the order does matter. The number of permutations of n things chosen r at a time is found using 

```
P(n, r) = n! / (n−r)!
```

```
import math

outcomes = 20**10
print("Number of possible outcomes in order:", outcomes)
```

(b) How many outcomes were possible in your simulation if the order did not matter?

A combination is a grouping of outcomes in which the orders do not matter.
The number of combinations of n things chosen r at a time is found using 


```
import math

outcomes = math.comb(20, 10)
print("Number of possible outcomes when order doesn't matter:", outcomes)
```
